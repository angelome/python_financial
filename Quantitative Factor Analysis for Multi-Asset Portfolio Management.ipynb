```python
"""
Multi Asset Factor Model

Factors:
- Interest rates
- Inflation
- Growth
"""

import pandas as pd
import yfinance as yf
from sklearn.linear_model import LinearRegression

assets=["SPY","TLT","GLD"]
prices=yf.download(assets,start="2015-01-01")["Close"]
returns=prices.pct_change().dropna()

# Example macroeconomic factors
factors=pd.DataFrame({
"Rate_Change":0.01,
"Inflation":0.02,
"Economic_Growth":0.03
},index=returns.index)

model=LinearRegression()
factor_results=[]

for asset in assets:
    model.fit(factors,returns[asset])
    factor_results.append({
    "Asset":asset,
    "Rate Sensitivity": model.coef_[0],
    "Inflation Sensitivity": model.coef_[1],
    "Growth Sensitivity": model.coef_[2],
    "Alpha": model.intercept_
    })

results=pd.DataFrame(factor_results)

print(results)


#results.to_excel("factor_analysis.xlsx",index=False)
```
