```python
"""
Fixed Income Pricing Engine

Supports:
- Fixed coupon bonds
- Zero coupon bonds
- Floating rate bonds
"""


import numpy as np

def fixed_rate_bond(face,coupon,yield_rate,years,frequency=2):
    periods=years*frequency
    coupon_payment=face*coupon/frequency
    discount=yield_rate/frequency
    pv=sum(coupon_payment/(1+discount)**t
    for t in range(1,periods+1))
    pv += face/(1+discount)**periods
    return pv

def zero_coupon(face,yield_rate,years):
    return face/(1+yield_rate)**years

def floating_rate_bond(face,spread,reference_rate,years):
    coupon=reference_rate+spread
    return fixed_rate_bond(face,coupon,reference_rate,years)

# modify the numbers below
print("Fixed Bond Price:",fixed_rate_bond(100,0.03,0.045,10))
print("Zero Coupon Price:",zero_coupon(100,0.04,5))
```
