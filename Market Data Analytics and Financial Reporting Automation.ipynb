```python
import pandas as pd
import yfinance as yf

# Define portfolio weights
portfolio = {
    "SPY": 0.40,
    "TLT": 0.40,
    "GLD": 0.20
}

# Download data and extract Close prices
prices = yf.download(list(portfolio.keys()), period="1y")["Close"]

# Calculate daily returns
returns = prices.pct_change()

# Calculate weighted portfolio return
portfolio_return = (returns * pd.Series(portfolio)).sum(axis=1)

# Calculate metrics for summary
avg_daily_ret = portfolio_return.mean()
ann_ret = (1 + avg_daily_ret)**252 - 1
ann_vol = portfolio_return.std() * (252**0.5)

# Create report DataFrame
# Using rolling(30) to show how volatility changes over time
report = pd.DataFrame({
    "Daily Return": portfolio_return,
    "Cumulative Return": (1 + portfolio_return).cumprod(),
    "Rolling Vol (30d)": portfolio_return.rolling(window=30).std() * (252**0.5)
})

# Create summary DataFrame
summary = pd.DataFrame({
    "Average Daily Return": [avg_daily_ret],
    "Annual Return": [ann_ret],
    "Annual Volatility": [ann_vol]
})

# Display Output
print("=" * 60)
print("INVESTMENT COMMITTEE REPORT")
print("=" * 60)

print("\nPORTFOLIO SUMMARY")
print(summary.round(4))

print("\nDAILY ANALYTICS (Last 10 Trading Days)")
print(report.tail(10).round(4))

print("\nKEY PERFORMANCE METRICS")
print(f"Average Daily Return : {avg_daily_ret:.4%}")
print(f"Annual Return        : {ann_ret:.2%}")
print(f"Annual Volatility    : {ann_vol:.2%}")
print(f"Cumulative Return    : {(report['Cumulative Return'].iloc[-1] - 1):.2%}")
print(f"Best Daily Return    : {portfolio_return.max():.2%}")
print(f"Worst Daily Return   : {portfolio_return.min():.2%}")

print("\nReport generated successfully.")
```
