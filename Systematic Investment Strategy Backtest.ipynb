```python
"""
Systematic Investment Strategy Backtest
"""

import pandas as pd
import numpy as np
import yfinance as yf

prices=yf.download("TLT",start="2010-01-01")["Close"]
returns=prices.pct_change()
moving_average=prices.rolling(200).mean()
signal=(prices>moving_average)
strategy_returns=(returns*signal.shift(1))
cumulative=(1+strategy_returns).cumprod()
sharpe=(strategy_returns.mean()/strategy_returns.std()*np.sqrt(252))
max_drawdown=(cumulative/cumulative.cummax()-1).min()

results=pd.DataFrame({
"Metric":["Sharpe Ratio","Maximum Drawdown","Total Return"],
"Value":[sharpe,max_drawdown,cumulative.iloc[-1]-1]})

print(results)

#results.to_excel("backtest_results.xlsx",index=False)
```
