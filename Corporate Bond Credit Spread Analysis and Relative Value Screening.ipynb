```python
"""
Corporate Bond Relative Value Model
"""

import pandas as pd

bonds=pd.DataFrame({
"Issuer":["Bank A","Industrial B","Energy C","Telecom D"],
"Rating":["AA","BBB","A","BBB"],
"Yield":[5.1,6.3,5.7,6.0],
"Duration":[4.2,6.5,5.0,7.1]})

sovereign_yield=3.5

bonds["Credit Spread"] = (bonds["Yield"]-sovereign_yield)
bonds["Spread/Duration"] = (bonds["Credit Spread"]/bonds["Duration"])
ranking=bonds.sort_values("Spread/Duration",ascending=False)

print(ranking)

# Attractive opportunities
screen = ranking[(ranking["Credit Spread"]>2)&(ranking["Rating"]!="AA")]
print("\nRelative Value Opportunities")
print(screen)


#ranking.to_excel("credit_screening.xlsx",index=False)
```
