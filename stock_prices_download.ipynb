```python
#import packages to use in this script
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import datetime as dt
import seaborn as sns
sns.set()
import yfinance
```

```python
# define stock/index of interest, start and end date of the time series
stock = "^GSPC"
start_date = dt.datetime(2020, 12, 31)
end_date = dt.datetime(2021, 12, 31)
raw_data = yfinance.download (tickers = stock, start = start_date, end = end_date, interval = "1d") 
```

```python
# plot the time series
plt.figure(figsize = (10,5))
plt.title('Prices from {} to {}'.format(start_date, end_date))
plt.plot(raw_data['Adj Close'])
plt.show()
```
